```python
from simple_esn.ESN_Ridge_learner import ESN_Ridge_learner
```

```python
from numpy import loadtxt, atleast_2d
if __name__ == '__main__':
    X = loadtxt('/anaconda3/lib/python3.7/site-packages/simple_esn/MackeyGlass_t17.txt')
    X = atleast_2d(X).T
    train_length = 2000
    test_length = 2000

X_train = X[:train_length]
y_train = X[1:train_length+1]
X_test = X[train_length:train_length+test_length]
y_test = X[train_length+1:train_length+test_length+1]

esn = ESN_Ridge_learner(n_readout=100)
model = esn.fit(X_train,y_train)
model.predict(X_train)
```

```python
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format='retina'

df = pd.DataFrame({'true': X_train.ravel().tolist(),
'predict': model.predict(X_train).ravel().tolist()})
df.plot(title='Train Data')
```

```python
df = pd.DataFrame({'true': X_test.ravel().tolist(),
'predict': model.predict(y_test).ravel().tolist()})
df.plot(title='Test Data')
```
