# 模型预测

```python
import sys
sys.path.append('/Users/apple/Documents/ML_Project/ML - 2.1/')
import numpy as np
import pandas as pd
from module.utils import *
from ngboost.learners import *
from tqdm.notebook import tqdm as tqdm
%config InlineBackend.figure_format='retina'
```

```python
transform='sin+cos'
X_train, X_test, Y_train, Y_test, Y_scaler = get_data2(hour_num=0, transform=transform,
                                                       drop_time=True, drop_else=False,
                                                       scale=True, return_y_scaler=True)
Pred_df = Y_test
```

## default_linear_learner

```python
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt
mse_list = []
for alpha in tqdm(np.arange(0,1,0.01)):
    Y_pred = Ridge(alpha=alpha).fit(X_train, Y_train).predict(X_test)
    mse = mean_squared_error(Y_pred, Y_test)
    mse_list.append(mse)
Series = pd.Series(mse_list, index=np.arange(0,1,0.01), name='mse')
Series.plot(title='MSE')
show_min = '['+str(Series.idxmin())+' '+str(round(Series.min(), 4))+']'
plt.annotate(show_min,xytext=(Series.idxmin(),Series.min())
             ,xy=(Series.idxmin(),Series.min()))
print('best alpha:', Series.idxmin())
```

```python
Y_pred = model_test(Base=default_linear_learner(alpha=1),
           X_train=X_train, X_test=X_test,
           Y_train=Y_train, Y_test=Y_test,
          n_estimators=500, verbose_eval=100,
          plot_predict=True, return_y_pred=True,
          Y_scaler=Y_scaler)
Y_pred.name = 'default_linear_learner'
Pred_df = pd.concat([Pred_df, Y_pred], axis=1)
pd.Series(np.zeros(len(Pred_df)), index=Pred_df.index).plot(color='k')
del Y_pred
```

## default_tree_learner

```python
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
mse_list = []
for depth in tqdm(np.arange(1,20,1)):
    Y_pred = DecisionTreeRegressor(max_depth=depth).\
    fit(X_train, Y_train).predict(X_test)
    mse = mean_squared_error(Y_pred, Y_test)
    mse_list.append(mse)
Series = pd.Series(mse_list, index=np.arange(1,20,1), name='mse')
Series.plot(title='MSE')
show_min = '['+str(Series.idxmin())+' '+str(round(Series.min(), 7))+']'
plt.annotate(show_min,xytext=(Series.idxmin(),Series.min())
             ,xy=(Series.idxmin(),Series.min()))
print('best depth:', Series.idxmin())
Series

Y_pred = DecisionTreeRegressor(max_depth=Series.idxmin()).\
    fit(X_train, Y_train).predict(X_test)
mse = mean_squared_error(Y_pred, Y_test)
pd.concat([pd.Series(Y_pred, name='Pred', 
                     index=Y_test.index), Y_test], axis=1)\
.plot(title='mse:'+str(round(mse, 8))+'   depth:'+str(Series.idxmin()))
```

```python
Y_pred = model_test(Base=default_tree_learner(depth=10),
           X_train=X_train, X_test=X_test,
           Y_train=Y_train, Y_test=Y_test,
          n_estimators=500, verbose_eval=200,
          plot_predict=True, return_y_pred=True,
          Y_scaler=Y_scaler)
Y_pred.name = 'default_tree_learner'
Pred_df = pd.concat([Pred_df, Y_pred], axis=1)
pd.Series(np.zeros(len(Pred_df)), index=Pred_df.index).plot(color='k')
del Y_pred
```

## lasso_learner

```python
from sklearn.linear_model import Lasso
import matplotlib.pyplot as plt
mse_list = []
for alpha in tqdm(np.arange(0,1,0.01)):
    Y_pred = Lasso(alpha=alpha).fit(X_train, Y_train).predict(X_test)
    mse = mean_squared_error(Y_pred, Y_test)
    mse_list.append(mse)
Series = pd.Series(mse_list, index=np.arange(0,1,0.01), name='mse')
Series.plot(title='MSE')
show_min = '['+str(Series.idxmin())+' '+str(round(Series.min(), 4))+']'
plt.annotate(show_min,xytext=(Series.idxmin(),Series.min())
             ,xy=(Series.idxmin(),Series.min()))
print('best alpha:', Series.idxmin())
```
