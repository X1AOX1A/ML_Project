```python
def tree_model_plot2(best_param, drop_minute=False, 
                     test_index=[2000, 3001],
                     filename='20738-2012.csv'):
    path='~/Documents/ML_Project/ML - 2.1/Data/相近8个地点2012年数据/'+filename
    X_train, X_test, Y_train, Y_test, Y_scaler = \
    get_data2(hour_num=int(best_param['hour_num']),
             transform=best_param['transform'],
             drop_time=bool(best_param['drop_time']),
             drop_else=bool(best_param['drop_else']),
             scale=True, return_y_scaler=True,
             test_index=test_index,
             drop_minute=drop_minute, path=path)
    Y_pred = DecisionTreeRegressor(max_depth=best_param['max_depth']).\
        fit(X_train, Y_train).predict(X_test)
    mse = mean_squared_error(Y_pred, Y_test)
    pd.concat([pd.Series(Y_pred, name='Pred', 
                         index=Y_test.index), Y_test], axis=1)\
    .plot(title='mse:'+str(round(mse, 8))+\
          '   depth:'+str(best_param['max_depth'])+\
          '   file:'+filename.split('.')[0]+\
          '   test_index:'+str(test_index),
          figsize=(12, 5))
    plt.savefig('/Users/apple/Documents/ML_Project/ML - 2.1/result/Tree_model/figure_decision_tree_2/'+str(test_index)+filename.split('.')[0]+'.png')
```

```python
best_param =  {'drop_else': 1.0, 
               'drop_time': 1.0, 
               'hour_num': 1.0, 
               'max_depth': 10.0, 
               'mse': 2.5709866985355128e-05, 
               'transform': 'sin+cos'} 

files = [
'20738-2012.csv',
'20739-2012.csv',
'20874-2012.csv',
'20875-2012.csv',
'20876-2012.csv',
'21016-2012.csv',
'21017-2012.csv',
'21018-2012.csv']

for test_index in [[0,1001], [1000,2001], [2000,3001]]:
    for filename in files:
        tree_model_plot2(best_param, drop_minute=False, 
                         filename=filename, test_index=test_index)
```
